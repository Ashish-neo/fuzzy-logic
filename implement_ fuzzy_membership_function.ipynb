```python
import numpy as np
import skfuzzy as fuzzy
from skfuzzy import control as ctrl 
```

```python
import matplotlib.pyplot as plt
abc=ctrl.Antecedent(np.arange(1,10,1),"abc")
abc['Triangular']=fuzzy.trimf(abc.universe,[3,4,6])
abc.view()
```

```python
tra=ctrl.Antecedent(np.arange(1,41,1),"tra")
tra['Trapezoidal']=fuzzy.trapmf(tra.universe,[13,20,25,28])
tra.view()
```

```python
ms=ctrl.Antecedent(np.arange(0.4,10,0.5),"ms")
ms['Sigmoid']=fuzzy.sigmf(ms.universe,4,6)
ms.view()
```

```python
abc=ctrl.Antecedent(np.arange(0.4,10,0.5),"abc")
abc['Gaussian']=fuzzy.gaussmf(abc.universe,5,2)
abc.view()
```

```python
abc=ctrl.Antecedent(np.arange(0,10,0.2),"abc")
abc['Generalized Bell']=fuzzy.gbellmf(abc.universe,2,4,6)
abc.view()
```
